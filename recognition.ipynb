```python
def faceencodings(images):
    encodelist=[]
    for i in img:
        img1=cv2.cvtColor(i,cv2.COLOR_BGR2RGB)
        face_in_frame=fr.face_locations(img1)
        print(face_in_frame)
        face_encode=fr.face_encodings(img1,face_in_frame)[0]
        encodelist.append(face_encode)
    return encodelist    
```

```python
faceencodings(img)[0]
```

```python
encodelist_knownface=faceencodings(img)
```

```python
video=cv2.VideoCapture(0)
while True:
    suc,img=video.read()
    img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    face_loc1=fr.face_locations(img1)
    #print(face_loc1)
    faceencode1=fr.face_encodings(img1,face_loc1)
    for enc_loc,faceloc in zip(faceencode1,face_loc1):
        matches=fr.compare_faces(encodelist_knownface,enc_loc)
        print(matches)
        face_dis=fr.face_distance(encodelist_knownface,enc_loc)
        print(face_dis)
        matchindex=np.argmin(face_dis)
        print(matchindex)
        if matches[matchindex]==True:
            name=classnames[matchindex]
            print(name)

           
            txt_sp=pyttsx3.init()
            voices=txt_sp.getProperty('voices')
            txt_sp.setProperty('voice',voices[1].id)
            txt_sp.setProperty('volume',5)
            txt_sp.setProperty('rate',150)

            txt_sp.say(name)
            txt_sp.runAndWait()

            y1,x2,y2,x1=faceloc
            cv2.rectangle(img,(x1,y1),(x2,y2),(0,0,255),3)
            cv2.putText(img,name,(x1,y1),cv2.FONT_HERSHEY_COMPLEX,1,(0,0,255),3)
        else:
            txt_sp=pyttsx3.init()
            voices=txt_sp.getProperty('voices')
            txt_sp.setProperty('voice',voices[1].id)
            txt_sp.setProperty('volume',5)
            txt_sp.setProperty('rate',150)
            name='unknown'
            txt_sp.say(name)
            txt_sp.runAndWait()
             
            y1,x2,y2,x1=faceloc
            cv2.rectangle(img,(x1,y1),(x2,y2),(0,0,255),3)
            cv2.putText(img,"No match",(x1,y1),cv2.FONT_HERSHEY_COMPLEX,1,(0,0,255),3)
             


    cv2.imshow('Frame',img)
    if cv2.waitKey(1)&0XFF==ord('q'):
        break
video.release()
cv2.destroyAllWindows()    
```
